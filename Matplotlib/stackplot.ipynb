```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]

# Default stackplot
plt.figure(figsize=(8, 6))
plt.stackplot(x, area1, area2, area3, labels=['area1', 'area2', 'area3'])
plt.title('Default Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]

# Normalized stackplot
plt.figure(figsize=(8, 6))
plt.stackplot(x, area1, area2, area3, labels=['area1', 'area2', 'area3'], normalize=True)
plt.title('Normalized Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]

# Offset stackplot
plt.figure(figsize=(8, 6))
plt.stackplot(x, area1, area2, area3, labels=['area1', 'area2', 'area3'], baseline='zero', baseline_offset='wiggle')
plt.title('Offset Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]

# Percentage stackplot
plt.figure(figsize=(8, 6))
plt.stackplot(x, area1, area2, area3, labels=['area1', 'area2', 'area3'], baseline='zero', normalize='percent')
plt.title('Percentage Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

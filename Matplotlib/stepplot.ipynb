```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 10, 1)
y = np.array([1, 3, 2, 5, 7, 8, 6, 10, 12, 9])

# Plot
plt.step(x, y)
plt.title('Default Step Plot (pre)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 10, 1)
y = np.array([1, 3, 2, 5, 7, 8, 6, 10, 12, 9])

# Plot
plt.step(x, y, where='mid', linestyle='--', color='r', marker='o')
plt.title('Customized Step Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 10, 1)
y1 = np.array([1, 3, 2, 5, 7, 8, 6, 10, 12, 9])
y2 = np.array([2, 2, 3, 4, 6, 6, 5, 8, 10, 7])

# Plot
plt.step(x, y1, where='pre', label='Line 1')
plt.step(x, y2, where='post', label='Line 2')
plt.title('Multiple Lines Step Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python","c","c++","java"]
y = [85,70,60,82]
plt.xlabel("Language",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.title("BarPlot",fontsize=15)
c=["y","b","g","m"]
plt.bar(x,y,width=0.4,color=c,label="popularity",alpha=0.4,align="edge",edgecolor="r",linewidth=5,linestyle=":")
plt.legend(loc=9,facecolor="Red")
plt.show()
```

```python
import numpy as np

x = ["python","c","c++","java"]
y = [85,70,60,82]
z=[20,30,40,50]

width = 0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.title("BarPlot",fontsize=15)

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 2, 5]
values2 = [4, 6, 3, 8]

# Plotting
plt.bar(categories, values1, color='blue', label='Series 1')
plt.bar(categories, values2, bottom=values1, color='red', label='Series 2')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 2, 5]
values2 = [4, 6, 3, 8]

# Parameters
bar_width = 0.35
index = np.arange(len(categories))

# Plotting
plt.bar(index, values1, bar_width, color='blue', label='Series 1')
plt.bar(index + bar_width, values2, bar_width, color='red', label='Series 2')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Grouped Bar Plot')
plt.xticks(index + bar_width / 2, categories)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]
errors = [0.5, 1.0, 0.2, 0.7]

# Plotting
plt.bar(categories, values, yerr=errors, color='purple', capsize=5)
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Bar Plot with Error Bars')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]
colors = ['red', 'blue', 'green', 'orange']

# Plotting
plt.bar(categories, values, color=colors)
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Bar Plot with Custom Colors')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]
patterns = ['/', '\\', '|', '-']

# Plotting
bars = plt.bar(categories, values, color='skyblue')
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)

plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Bar Plot with Patterns')
plt.show()
```

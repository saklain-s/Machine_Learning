```python
import matplotlib.pyplot as plt

# Data
data1 = [20, 41, 32, 10, 22, 12, 43, 40, 30, 27, 10, 36, 37, 36, 21, 33, 48, 
         16, 22, 40, 48, 21, 34, 49, 48, 34, 10, 47, 14, 16]
data2 = [20, 15, 30, 25, 22, 18, 35, 32, 28, 27, 25, 36, 33, 34, 21, 30, 48, 
         16, 22, 40, 48, 21, 34, 45, 48, 34, 10, 47, 14, 16]

# Plotting
plt.hist(data1, bins=10, density=True, alpha=0.5, color='blue', edgecolor='black', label='Data 1')
plt.hist(data2, bins=10, density=True, alpha=0.5, color='red', edgecolor='black', label='Data 2')
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Normalized Histogram')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
data1 = [20, 41, 32, 10, 22, 12, 43, 40, 30, 27, 10, 36, 37, 36, 21, 33, 48, 
         16, 22, 40, 48, 21, 34, 49, 48, 34, 10, 47, 14, 16]
data2 = [20, 15, 30, 25, 22, 18, 35, 32, 28, 27, 25, 36, 33, 34, 21, 30, 48, 
         16, 22, 40, 48, 21, 34, 45, 48, 34, 10, 47, 14, 16]

# Plotting
plt.hist([data1, data2], bins=10, stacked=True, color=['blue', 'red'], edgecolor='black', label=['Data 1', 'Data 2'])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data (exponential distribution)
data = np.random.exponential(scale=1.0, size=1000)

# Plotting
plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.yscale('log')  # Logarithmic scale on y-axis
plt.xlabel('Values')
plt.ylabel('Frequency (log scale)')
plt.title('Histogram with Log Scale')
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["r","y","g","b","r","g","r"]
s=[400,200,400,300,200,1100,600]

plt.xlabel("No",fontsize=15)
plt.ylabel("Days",fontsize=15)
plt.title("Scatter-Plot",fontsize=15)


plt.scatter(day,no,color=c,sizes=s,alpha=0.4,marker="o")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)
colors = np.linspace(0, 1, 100)

# Plotting with different colormaps
colormaps = ['viridis', 'plasma', 'inferno', 'magma', 'cividis']

for i, cmap in enumerate(colormaps):
    plt.subplot(1, len(colormaps), i + 1)
    plt.scatter(x, y, c=colors, cmap=cmap)
    plt.title(cmap)
    plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

# Contour plot with colormap
plt.contourf(X, Y, Z, cmap='plasma')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# Data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

# Custom colormap
custom_cmap = LinearSegmentedColormap.from_list('custom_cmap', ['blue', 'white', 'red'])

# Plotting
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap=custom_cmap)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
labels = ['A', 'B', 'C', 'D', 'E']

# Plotting
plt.scatter(x, y, color='blue')
for i, label in enumerate(labels):
    plt.text(x[i], y[i], label, fontsize=12, ha='right')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Labels')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting
plt.scatter(x, y, color='blue')
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0, 10), ha='center')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Annotations')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

# Plotting
plt.scatter(x, y, color='blue')

# Trend line
m, b = np.polyfit(x, y, 1)  # Fit line
plt.plot(x, m*x + b, color='red')  # Plot line

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Trend Line')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [11, 7, 5, 3, 2]

# Plotting
plt.scatter(x1, y1, color='blue', label='Series 1')
plt.scatter(x2, y2, color='red', label='Series 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Multiple Series')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z = [1, 4, 9, 16, 25]

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='blue')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')
plt.show()
```

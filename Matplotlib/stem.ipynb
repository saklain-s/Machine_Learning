```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt=":", markerfmt="ro", basefmt="g", label="Data")
plt.scatter(x, y, color='r', marker='o', facecolors='none', edgecolors='r', s=100)  # Adding scatter plot for filled markers
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]
markers = ['o', 's', '^', 'd', '>', '<']

for i in range(len(x)):
    plt.stem(x[i:i+1], y[i:i+1], linefmt=":", markerfmt=markers[i] + 'r', basefmt="g", label="Data")

plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y1 = [2, 2, 5, 6, 4, 3]
y2 = [1, 3, 4, 5, 2, 4]

plt.stem(x, y1, linefmt=":", markerfmt="ro", basefmt="g", label="Data 1")
plt.stem(x, y2, linefmt="--", markerfmt="bs", basefmt="y", label="Data 2", bottom=0)
plt.legend()
plt.show()
```

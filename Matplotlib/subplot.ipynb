```python
import matplotlib.pyplot as plt

# Create 1x2 grid of subplots
plt.figure()

plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title('Subplot 1')

plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [1, 2, 3])
plt.title('Subplot 2')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)  # 2x2 grid

axs[0, 0].plot([1, 2, 3], [1, 4, 9])
axs[0, 0].set_title('Subplot 1')

axs[0, 1].plot([1, 2, 3], [1, 2, 3])
axs[0, 1].set_title('Subplot 2')

axs[1, 0].plot([1, 2, 3], [1, 1, 1])
axs[1, 0].set_title('Subplot 3')

axs[1, 1].plot([1, 2, 3], [3, 2, 1])
axs[1, 1].set_title('Subplot 4')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure()

gs = gridspec.GridSpec(3, 3)

ax1 = fig.add_subplot(gs[0, :])
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title('Subplot 1: spans all columns')

ax2 = fig.add_subplot(gs[1, :-1])
ax2.plot([1, 2, 3], [1, 2, 3])
ax2.set_title('Subplot 2: spans first 2 columns')

ax3 = fig.add_subplot(gs[1:, -1])
ax3.plot([1, 2, 3], [1, 1, 1])
ax3.set_title('Subplot 3: spans last column')

ax4 = fig.add_subplot(gs[2, 0])
ax4.plot([1, 2, 3], [3, 2, 1])
ax4.set_title('Subplot 4')

ax5 = fig.add_subplot(gs[2, 1])
ax5.plot([1, 2, 3], [2, 3, 4])
ax5.set_title('Subplot 5')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title('Subplot 1: spans all columns')

ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax2.plot([1, 2, 3], [1, 2, 3])
ax2.set_title('Subplot 2: spans first 2 columns')

ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax3.plot([1, 2, 3], [1, 1, 1])
ax3.set_title('Subplot 3: spans last column')

ax4 = plt.subplot2grid((3, 3), (2, 0))
ax4.plot([1, 2, 3], [3, 2, 1])
ax4.set_title('Subplot 4')

ax5 = plt.subplot2grid((3, 3), (2, 1))
ax5.plot([1, 2, 3], [2, 3, 4])
ax5.set_title('Subplot 5')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Main plot
ax.plot([1, 2, 3], [1, 4, 9])
ax.set_title('Main Plot')

# Inset plot
inset_ax = fig.add_axes([0.5, 0.5, 0.4, 0.4])  # [left, bottom, width, height]
inset_ax.plot([1, 2, 3], [1, 2, 3])
inset_ax.set_title('Inset Plot')

plt.show()
```

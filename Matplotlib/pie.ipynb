```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 30, 45, 10]  # Sizes in percentage

# Plotting the basic pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Basic Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 30, 45, 10]  # Sizes in percentage
explode = (0, 0.1, 0, 0)  # "Explode" the 2nd slice (Banana)

# Plotting the exploded pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Exploded Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 30, 45, 10]  # Sizes in percentage

# Plotting the donut chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
plt.axis('equal')
plt.title('Donut Chart')
# Draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Importing 3D plotting library

# Data to plot
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 30, 45, 10]  # Sizes in percentage

# Plotting the 3D pie chart
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')  # Adding a 3D subplot
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
ax.set_title('3D Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels_outer = ['Fruit', 'Vegetables']
sizes_outer = [60, 40]  # Sizes in percentage

labels_inner = ['Apple', 'Banana', 'Cherry', 'Potato', 'Carrot']
sizes_inner = [15, 30, 15, 20, 20]  # Sizes in percentage

# Plotting the nested pie chart
fig, ax = plt.subplots()
ax.pie(sizes_outer, labels=labels_outer, autopct='%1.1f%%', startangle=90)

# Draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

# Draw a smaller pie chart inside with the inner data
ax2 = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

ax.pie(sizes_inner, labels=labels_inner, autopct='%1.1f%%', startangle=90)
ax.set_title('Nested Pie Chart')
plt.axis('equal')
plt.show()
```
